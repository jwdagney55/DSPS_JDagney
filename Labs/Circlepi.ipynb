What you know: The ratio of the area of the circle to the area of the square is Ï€ / 4.

```
def square(l, ax):
  ax.plot([-l, l], [-l, -l], 'k-')
  ax.plot([-l, l], [l, l], 'k-')
  ax.plot([-l, -l], [-l, l], 'k-')
  ax.plot([l, l], [-l, l], 'k-')
  
```

```
def circle(l, ax):
  theta = np.linspace(0, 360, 100)
  theta *= u.degree
  x = l * np.cos(theta.to(u.rad))
  y = l * np.sin(theta.to(u.rad))
  ax.plot(x, y, 'k-')
  
```

```
fig, ax = pl.subplots(figsize=(5, 5))
square(1, ax=ax)
circle(1, ax=ax)
```

```
def isin(x, y, l):
  #return ( ( ( l* np.cos(x)) ** 2 + ( (l * np.sin(y) )** 2) <= l ** 2 )
  return (  (l * x)**2 + (l * y )**2 <= l**2   )
    
```
